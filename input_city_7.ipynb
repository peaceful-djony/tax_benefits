```python
import pandas as pd
```

```python
data = pd.read_excel('Датасет от Ростелеком.xlsx',
                     skiprows=1,
                     usecols='A:L')
data=pd.DataFrame(data)
data.head()
```

```python
#добавили столбец, в котором названия регионов будут в нижнем регистре
data['Name_lower_primary'] = data['Субъект Российской Федерации'].str.lower()
# делаем столюбец без области, края и т.п.
data['Name_lower'] = data['Name_lower_primary'].str.replace('область', '').str.replace('республика', '').str.replace('автономный', '').str.replace('край', '').str.replace('округ', '').str.strip()
data.head()
```

```python
# список регионов в нижнем регистре
cities = data['Name_lower'].tolist()
cities
```

```python
# вводим название региона и сразу переводим в нижний регистр
word = input("Введите название региона (крайя, автономного округа, области или город Москва): ").lower()

# проходим по каждому городу из списка
for city in cities:
    city = str(city)
    num_letters = len(city)
    match_count = sum(1 for c1, c2 in zip(city, word) if c1 == c2)
    # если все буквы совпадают
    if match_count == num_letters:
        target_row=data[data['Name_lower'] == city]. index [ 0 ]
        target_row = int(target_row)
        target_col_tax_deduction=data.columns.get_loc('Закон ИНВ')
        cell_value = data.iat[target_row,target_col_tax_deduction]
        # чтобы наш нижний регистр не был виден, сделаем все большими буквами
        city_upper=city.upper()
        print(f"Вы выбрали регион {city_upper}. Закон ИНВ:{cell_value}.")
        break
    # если более половины букв совпадают
    elif match_count > num_letters / 2:
        
        city_upper=city.upper()
        print(f"Вы имели в виду {city_upper}? ТУТ НУЖНО ВЫБИРАТЬ ДА/НЕТ")
        break
# если совсем ничего не совпадает
else:
    print("Название региона не найдено. Попробуйте ввести город без опечаток и  аббревиатур. Возможно, вашего региона нет в списке")
```
